```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [21,345,56,12,5,47,1]
plt.plot(y)
```

```python
plt.figure(figsize=(15,5))
plt.pie(y)
plt.show()
```

```python
fig,ax = plt.subplots(nrows=2, ncols=1, figsize=(15,5) )
plt.show()
```

```python
t = np.arange(-3 * np.pi, 3 * np.pi, 0.2)
x = np.sin(t)
y = np.cos(t)
fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(10,6))
ax[0].plot(t,x,'r')
ax[1].bar(t,y)
plt.show()
```

```python
x = [100,200,300,400,500,600,700,800,900,1000]
y = [23,45,1,34,65,2,2,46,34,56]
plt.plot(x, y, linewidth=2, color='r', marker='o', 
markersize=15, markerfacecolor='white', markeredgecolor='green', linestyle='--')
```
